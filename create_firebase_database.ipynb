```python
from datetime import datetime

import locale
locale.setlocale(locale.LC_ALL, ("es_ES", "UTF-8"))
print(datetime.now().strftime("%H:%M %d/%b/%Y"))

año = 2010
df = pd.read_csv('la pepita_data.csv')
df.periodo = df.periodo.apply(lambda periodo: datetime.strptime(periodo, '%Y-%m-%d'))
df = df[df.periodo.dt.year == año].groupby(df.periodo.dt.month)['mm_lluvia'].sum().reset_index()
df.periodo = df.periodo.apply(lambda mes: datetime.strptime(str(mes), '%m').strftime("%B").capitalize())
df.periodo.to_list()
df.mm_lluvia.to_list()
```
